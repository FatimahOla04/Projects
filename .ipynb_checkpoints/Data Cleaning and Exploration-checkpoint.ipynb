# Data Cleaning

```python
# fill the missing value in age with median
train["Age"].median()
```

```python
age_mea=train["Age"].median()
train["Age"].fillna(age_mea, inplace=True)
```

```python
# fill the missing value in cabin with mode
cabin_mode=train["Cabin"].mode()[0]
train["Cabin"].fillna(cabin_mode,inplace=True)
```

```python
# fill the missing value in Embarked with mode
Embarked_mode=train["Embarked"].mode()[0]
train["Embarked"].fillna(Embarked_mode,inplace=True)
```

```python
# drop all the duplicates
train.drop_duplicates(inplace=True)
```

```python
train["Name"]
```

```python
for i in train["Name"]:
    print(i.split()[1].strip(".").strip(","))
```

```python
# Create a new column title in the reain
t=[]
for i in train["Name"]:
    t.append(i.split()[1].strip(".").strip(","))
train["Title"]=t
```

```python
train
```

```python
train["Title"].value_counts()
```

```python
major_titles=["Mr", "Miss", "Mrs", "Dr", "Rev", "Master"]
for i in train["Title"]:
    if i not in major_titles:
        train["Title"].replace(i,"other", inplace=True)
```

## Bi-variate Analysis: comparing two columns to understand the patterns between them

- cat vs cat
- cat vs cons
- cons vs cons

```python
pd.crosstab(index=train["Survived"], columns=train["Pclass"], values=train["Pclass"], aggfunc="count")
```

```python
def barplot(col):
    u=pd.crosstab(columns=train["Survived"], index=train[col], values=train[col], aggfunc="count")
    fig, ax = plt.subplots(figsize = (6, 6), facecolor="black")
    ax.set_facecolor("black")
    u.plot(kind="bar", ax=ax)
    plt.title(f"Count of Survival by {col}", color="red")
    plt.ylabel("Count")
    plt.xlabel("Title")
    plt.xticks(rotation=35, color="green")
    plt.yticks(rotation=35, color="green")
    plt.show
```

```python
barplot("Sex")
```

```python
# Category vs continuous
```

```python
# To find the average of sex and age, we need to group them by using groupby
train.groupby("Sex").agg({"Age":"mean"}).round().plot(kind="bar")
fig, ax = plt.subplots(figsize = (6, 6), facecolor="black")
ax.set_facecolor("black")
plt.xticks(rotation=35, color="green")
plt.yticks(rotation=35, color="green")
plt.show()
```

## FacetGrid in Seaborn is used to create a grid of plots based on the values of one or more categorical variables. This allows for easy comparison of distributions or relationship across different subsets of the data

```python
ola=sns.FacetGrid(data=train, hue="Survived", aspect=4)
ola.map(sns.kdeplot, "Age", shade=True)
plt.xlim(0, 80)
plt.legend()
```
